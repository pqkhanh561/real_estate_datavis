## Clean data

```python
# diện tích đất 
df['Diện tích đất'] = df['Diện tích đất'].apply(lambda x: float(re.findall('[\d\.\d]+',x)[0]))
df['Diện tích đất'] # m²	
```

```python
# diện tích sử dụng
def price4use(text):
    res=''
    if pd.isnull(text):
        res =np.nan
    else:
        res = float(re.findall('[\d\.\d]+',text)[0])
    return res

df['Diện tích sử dụng'] = df['Diện tích sử dụng'].apply(lambda x: price4use(x))
df['Diện tích sử dụng'] # m²	
```

```python
df['Đơn giá gộp']
```

```python
df['Đơn giá gộp'] = df['Đơn giá gộp'].apply(lambda x: float(re.findall('[\d\,\d]+',x)[0].replace(',','.')))
df['Đơn giá gộp'] # triệu/m²
```

```python
df['Giá']
```

```python
def get_price(text): # đơn vị: tỷ
    res =''
    ty = re.search('[\d\,\d]+.tỷ$',text)
    trieu = re.search('[\d\,\d]+.triệu$',text)
    if ty:
        res = ty.group()
        res = re.findall('[\d\,\d]+',res)[0].replace(',','.')
        res = float(res)
    else:
        res = trieu.group()
        res = float(re.findall('[\d\,\d]+',res)[0])/1000
    
    return res
# 
df['Giá'] = df['Giá'].apply(lambda x: get_price(x))
df['Giá']
```

```python
df['Địa chỉ']
```

```python
df['Đường'] = df['Địa chỉ'].apply(lambda x: x.split(',')[0].strip())
df['Đường'] 
```

```python
def f(text): # chiều dài, chiều rộng, độ rộng hẻm , độ rộng mặt tiền đường
    res =''
    try:
        res = float(re.findall('[\d\.\d]+',text)[0])
    except:
        res = np.nan
    return res 
```

```python
def lau(text):
    res = ''
    try:
        lau = re.search('[\d]+.lầu', text)
        res = lau.group()
        res = int(re.findall('[\d]+',res)[0])
    except:
        res = np.nan
    return res
# lửng 
def lung(text):
    res = ''
    try:
        lung = re.search('[\d]+.lửng', text)
        res = lung.group()
        res = int(re.findall('[\d]+',res)[0])
    except:
        res = np.nan
    return res
# tầng thượng
def tangthuong(text):
    res = ''
    try:
        tangthuong = re.search('[\d]+.tầng thượng', text)
        res = tangthuong.group()
        res = int(re.findall('[\d]+',res)[0])
    except:
        res = np.nan
    return res    
        
    
```

```python
df['Lầu'] = df['Kết cấu nhà'].apply(lambda x: lau(x)).fillna(0).astype(int)
df['Lầu'] 
```

```python
df['Lửng'] = df['Kết cấu nhà'].apply(lambda x: lung(x)).fillna(0).astype(int)
df['Lửng']
```

```python
df['Tầng thượng'] = df['Kết cấu nhà'].apply(lambda x: tangthuong(x)).fillna(0).astype(int)
df['Tầng thượng'] 
```

```python
hcmc_location = pd.read_excel('hcmc.xlsx')
hcmc_location
```

```python
hcm = pd.merge(hcmc_location, price_district, on='Quận').sort_values(by='Giá')
hcm
    
```

```python
import folium
import branca
```

```python
def fancy_html(row):
    i = row
    district = hcm['Quận'].iloc[i]                             
    price_unit = round(hcm['Đơn giá gộp'].iloc[i],2)  # đơn giá gộp trung bình (triệu/m2)
    price = round(hcm['Giá'].iloc[i],2)  # giá: (tỷ/ căn)                                                             
                           
    
    
    left_col_colour = "#2A799C"
    right_col_colour = "#C5DCE7"
    
    html = """<!DOCTYPE html>
<html>

    <table style="height: 100px; width: 300px;">
<tbody>
<tr>
<td style="background-color: """+ left_col_colour +""";"><span style="color: #ffffff;">Quận</span></td>
<td style="width: 200px;background-color: """+ right_col_colour +""";">{}</td>""".format(district) + """
</tr>
<tr>
<td style="background-color: """+ left_col_colour +""";"><span style="color: #ffffff;">Đơn giá gộp trung bình cho 1 m² </span></td>
<td style="width: 200px;background-color: """+ right_col_colour +""";">{} triệu/m²</td>""".format(price_unit) + """
</tr>
<tr>
<td style="background-color: """+ left_col_colour +""";"><span style="color: #ffffff;">Giá trung bình 1 căn </span></td>
<td style="width: 200px;background-color: """+ right_col_colour +""";">{} tỷ/căn</td>""".format(price) + """
</tr>
</tbody>
</table>
</html>
"""
    return html
```

```python

hcm_map = folium.Map(location=[10.762622, 106.660172], zoom_start=12) ## hcm_latitude = 10.762622, hcm_longtitude = 106.660172
# đánh dấu vị trí các quận HCMC 
districts = folium.map.FeatureGroup()
# đánh dấu cho các quận 
for lat, lng, in zip(hcm.Latitude, hcm.Longitude):
    districts.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=6, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
# add pop-up text to each marker on the map
for i in range(0, len(hcm)):
    if hcm['Đơn giá gộp'].iloc[i]< 100: # đơn giá gộp trung bình < 100 tr/m2: màu xanh
        color = 'blue'
    else:
        color = 'red'
    html = fancy_html(i)
    iframe = branca.element.IFrame(html=html,width=400,height=300)
    popup = folium.Popup(iframe,parse_html=True)
    folium.Marker([hcm['Latitude'].iloc[i], hcm['Longitude'].iloc[i]], popup=popup, icon=folium.Icon(color=color, icon='info-sign')).add_to(hcm_map) 

# add districts to map
hcm_map.add_child(districts)
```
